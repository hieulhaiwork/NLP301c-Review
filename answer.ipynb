```python
# Q1.
import re

pattern = r"\b[a-z]+\b"
s = "Today New York has sunny and rainy On The afternoon."

list = re.findall(pattern, s)
print(list)
```

```python
# Q2.
import nltk
nltk.download('reuters')
from nltk.probability import FreqDist
from nltk.corpus import reuters

text5 = reuters.raw('test/14826')

tokens = [word for word in text5.split(" ") if len(word) == 4]
fdist = FreqDist(tokens)
print(fdist.most_common())
```

```python
# Q2. 
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.probability import FreqDist
import re

text = "Hello, thank you very much. This car is nice and you are so good."

pattern = r"[^\w\d]"

tokens = [word for word in re.split(pattern, text) if word and word not in stopwords.words('English')]
freqdist = FreqDist(tokens)
print(freqdist.most_common())
```

```python
# Q4.
import nltk
import numpy as np
from nltk.tokenize import word_tokenize

text1 = "John lives in Canada"
text2 = "James lives in America, though he's not from here"

def cosine_sim(em1, em2):
    return (em1 @ em2)/(np.linalg.norm(em1)*np.linalg.norm(em2))

em_text1 = word_tokenize(text1)
em_text2 = word_tokenize(text2)

print(cosine_sim(em_text1, em_text2))
```
