```python
# Q1.

with open('sample.txt', 'r') as file:
    file_content = file.read()

import re

pattern_word = r"\s"
pattern_char = r"[^\s]"
pattern_sen = r"\."

print(len(re.split(pattern_word, file_content)))
print(len(re.split(pattern_char, file_content)))
print(len(re.split(pattern_sen, file_content)) - 1)
```

```python
doc = "This is a sample document. This document contains several sentences. We can extract bigrams from it."

from nltk.probability import FreqDist

doc_list = re.findall("\w+", doc)
bigram_list = [(doc_list[i], doc_list[i+1]) for i in range(len(doc_list) - 1)]
freqdist = FreqDist(bigram_list)
dict(freqdist)
```
