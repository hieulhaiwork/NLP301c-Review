```python
flowers = ["carnation", "pendulum", "petunias", "begonia", "receptacle", "hostas", "pelorism", "paperwhite"]

import re

pattern_a = r"^pe\w*"

print("a.")
print([flower for flower in flowers if len(re.findall(pattern_a, flower)) > 0])

print("b.")
print([flower for flower in flowers if len(flower) < 8])
```

```python
text = 'She received the news of the discovery with equanimity'

import re

pattern_2 = r"\b\w*[^\s]\b"

print(re.findall(pattern_2, text)[-2:])
```

```python
# Q3. 
text = 'Please contact us at contact@fpt.edu.vn for further information. You can also give feedback at feedback@fpt.edu.vn'

pattern_3 = r"\w+@\w+.\w+(?:.\w+)"

print(re.findall(pattern_3, text))
```

```python
# Q4.
input = ['table'] * 10 + ['chair'] * 9

import nltk
from nltk.probability import FreqDist

freqdist = FreqDist(input)
print([word for (word, freq) in freqdist.most_common()])
```
